```python
import pandas as pd
```

```python
df=pd.read_csv('bmw.csv')
df.head(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
##Handling missing values
df.isnull().any(axis=0)
```

```python
df.isnull().sum()
```

```python
##Filling the missing values with the mean of column
df['Sales_fillNA']=df['engineSize'].fillna(df['engineSize'].mean())
df.isnull().sum()
```

```python
df.dtypes
```

```python
##Renaming columns
df=df.rename(columns={'Sales_fillNA':'SF'})
df.head()
```

```python
##change datatypes
df['Value_new']=df['mileage'].fillna(df['mileage'].mean()).astype(float)
df.head()
```

```python
df['New']=df['tax'].apply(lambda x:x*2)
df.head()
```

```python
##Data aggregating and grouping

##Grouping one
grouped_mean=df.groupby('transmission')['mileage'].mean()
print(grouped_mean)
```

```python
##Grouping two
grouped_sum=df.groupby(['transmission','fuelType'])['mpg'].sum()
print(grouped_sum)
```

```python
## Aggregate multiple function
grouped_agg=df.groupby('transmission')['mpg'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
##Merging and joining DataFrame

df1=pd.DataFrame({ 'keys':['A','B','C'],'Data1':[1,2,3]})
df2=pd.DataFrame({'keys':['A','B','D'],'Data2':[153,532,324]})
```

```python
print(df1)
```

```python
print(df2)
```

```python
pd.merge(df1,df2,on='keys',how='inner')##inner for keys present in both df1 and df2
```

```python
pd.merge(df1,df2,on='keys',how='outer')##outer for keys both present and absent in  df1 and df2
```

```python
pd.merge(df1,df2,on='keys',how='left')
```

```python
pd.merge(df1,df2,on='keys',how='right')
```
