## Pandas-DataFrame and Series

Panda is a powerful data manupulation lib in python, widely used for data analysis and data cleaning.it provides two primary data structure:Series and DataFrame.A series is a one-dimensional array-like object,while a DataFrame is a two-dimesional, size-mutable, and potentially heterogenoes tabular data structure with labeled axes(rows and colums)

```python
import pandas as pd
```

```python
## Series
## A series is a one-dimensional array-like object that can hold any data type,It is similar to a colum in a table.

data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## Create a Series from dictionary

data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame 
## From dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Banglore','New York','Florida']
}
 
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
##Create a Data from a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Banglore'},
    {'Name':'John','Age':32,'City':'Banglore'},
    {'Name':'Bappy','Age':32,'City':'Banglore'},
    {'Name':'Jack','Age':32,'City':'Banglore'},
]

df=pd.DataFrame(data)
print(df)
print(type(df))
```
