```python
year=int(input("Enter the number"))

if year%4==0:

    if year%100!=0 and year%400!=0:
        print("The year is leap year")
    elif year%400==0:
        print("The year is leap year")
    else:
        print("The year is not a leap year")    
else:
    print("The year is not a leap year")       
        
```
