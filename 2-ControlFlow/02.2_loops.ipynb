```python
##while loop
count=5

while (count>0):
    print(count)
    count=count-1
     
```

```python
##Loop control statements

##break and contine(skip this iteration and move on with next)
for i in range(10):
    if i==5:
        break
    if(i==2):
        continue
    print(i)
##pass do nothing
```

```python
#Nested Loops
for i in range(3):
    for j in range(4):
        print(f"i:{i} and j:{j}")
 
```

```python
##Prime number between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
