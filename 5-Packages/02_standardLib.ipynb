```python
##Data Serialization
##Data serialization is the process of converting complex data structures or objects into a format that can be easily stored or transmitted,
#and later reconstructed back into the original form
import json
data={'name':'krish','age':25}

json_str=json.dumps(data)##for str
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
##csv

import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['krish',32])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
##Regular expression
import re
pattern=r'\d+'##\d+ matches a sequence of one or more digits
text='There are 123 apples 456'
match=re.search(pattern,text)
print(match.group())##group() returns the actual substring that matched the pattern.
```
