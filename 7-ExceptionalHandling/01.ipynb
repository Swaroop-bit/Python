Understanding Exception

Exception handling in python allows you to handle errors gracefully and take corrective actions withour stopping the execution of the program.This lesson will cover the basics of exceptions, including how to use try,except,else, and finally blocks.

```python
try:
   a=b
except NameError as ex:
   print(f"The variables has not been assigned {ex}")   
```

```python
##Exception is parent class from which other are exception are defined

try:
    result=1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter the denominator greater than 0")

except Exception as ex1:##Should be written last
    print(ex1)    
```

```python
try:
    num=int(input('Enter a number'))
    result=10/num
except ValueError:
        print("This is not a valid number")
except ZeroDivisionError:
      print("enter denominator greater than0") 
except Exception as e:
      print(e)             
```

```python
##try,except,else block
try:
    num=int(input('Enter a number'))
    result=10/num
except ValueError:
        print("This is not a valid number")
except ZeroDivisionError:
      print("enter denominator greater than 0") 
except Exception as e:
      print(e)   
else:
      print(f"the result is {result}")
```

```python
##try,except,else and finally
try:
    num=int(input('Enter a number'))
    result=10/num
except ValueError:
        print("This is not a valid number")
except ZeroDivisionError:
      print("enter denominator greater than 0") 
except Exception as e:
      print(e)   
else:
      print(f"the result is {result}")
finally:
      print("Execution completed")      
```

```python
## File handling and Exceptional Handling
try:
    file=open('example.txt')
    content=file.read()
    print(content)
except FileNotFoundError:
    print("The file does't exist")
except Exception as e:
    print(e)    
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print('file close')    
```
