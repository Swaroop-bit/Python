```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")    
```

```python
##Call this function
even_or_odd(4)
```

```python
##Variable length Arguments
## Positional and keywords arguments

##Poisitional arguments
def print_numbers(*args):
    for numbers in args:
        print(numbers)

##Keywords arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")  

##Combine
def print_all(*args,**kwargs):
    for val in args:
        print(f"Poisitional argumen:{val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")              
```

```python
print_numbers(2,3412,423,534,"Vidhi")
```

```python
print_details(name="Swaroop",age=32,countr="India")
```

```python
print_all(1,2,4,234,"Krish",nams="idk")
```
