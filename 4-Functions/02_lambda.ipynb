####Lambda
These are small anonymous functions defined using the lambda function.They can have any number of arguments but only one expression.They are commonly used for short operations or as arguments to higher-order functions.

```python
##Syntax
lambda arguments: expression
```

```python
addition=lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
even1=lambda num:num%2==0
even1(12)
```

```python
addition = lambda x,y,z:x+y+z
```

```python
##map()->applies a function to all items in a list
numbers=[1,2,3,4,5,6]
list(map(lambda x:x**2,numbers))
```
