```python
## Method Overiding
class Animal:
    def speak(self):
        return "Sound of the animal"
    
## Derived Class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"    
    
## Derived Class
class Cat(Animal):
    def speak(self):
        return "Meow!"    
##Function that demonstrates polymorphism
        
```

```python
##Polymorphism with Functions and methods

class Shape:
    def area(self):
        return "The area of the figure is not defined"
    
class Reactangle(Shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth   

    def area(self):
        return self.length*self.breadth  

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius   

    def area(self):
        return self.radius*self.radius*3.14

def print_area(shape):
    print(f"The area is {shape.area()}")    

c1=Circle(10)
print_area(c1)
```
