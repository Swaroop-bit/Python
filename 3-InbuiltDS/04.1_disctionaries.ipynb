##Dictionaries

these are unordered collection of items.They store data in key-value pairs.Keys must be unique and immutable(e.g strings,numbers or tuples),while values can be of any type

```python
##Creating dictionaries
empty_dict={}
print(type(empty_dict))
empty_dict=dict()
```

```python
##name.age,grade->key
##krish,32,24->value
student={"name":"krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
##single key is always used
student={"name":"krish","age":32,"name":24}
```

```python
##accessing Dictionary Elements
##1st method
student={"name":"krish","age":32,"grade":24}
print(student['grade'])
print(student['age'])

##2nd method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not available"))##giving default value

```

```python
##Modifing Dictionary Elements
##Dictoonary are mutable,so u can dd,update or delete elements

##update and add
student["address"]="India"
print(student)

##delete
del student['grade']
print(student)
```

```python
##Dicitionary methods

keys=student.keys()##get all the key
print(keys)
values=student.values()##get all values
print(values)
items=student.items()##get all the key values pairs
print(items)
```

```python
##Shallow copy

##dedicate same memory so chnging one will chnge other
student_copy=student
student["name"]="Swaroop"
print(student)
print(student_copy)


##that's why we use shallow
student_copy2=student.copy()
```

```python
##Iterating over Dictionaries
##You can use loops to iterate over dictionarise,keys,values or items

##Iterating over keys
for keys in student.keys():
    print(keys)

##Iterate over values
for values in student.values():
    print(values)    

##Iterating over key values pairs
  
```

```python
##Nested
students={
    "student1":{"names":"Krish","age":32},
    "student2":{"names":"peter","age":35}
}
print(students)
```

```python
#Access nested dictionaries elements
print(students["student2"]["names"])
```

```python
##Iterating over nested
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for x,y in student_info.items():
        print(f"{x}:{y}")
        
```

```python
##Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
##Conditonal dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
##Practical Examples

##Use a dictionary to count the frquency of elements in list
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for x in numbers:
    if x in frequency:
        frequency[x]+=1
    else:
        frequency[x]=1

print(frequency)
```

```python
##Mergre 2 dictionaries

d1={"a":1,"b":2}
d2={"c":3,"d":4}
merge={**d1,**d2}
print(merge)
```
