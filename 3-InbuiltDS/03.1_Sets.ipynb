####Sets

Used to store collections of unique items.They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are use 

```python
##create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_set=set([1,2,3,6,5,4,5,6])
print((my_set))
```

```python
##Basic Sets Operation
##Adding and Removing Elements
my_set.add(3)
print(my_set)

my_set.add(7)
print(my_set)
```

```python
##Remove the elements
my_set.remove(2)
print(my_set)

##if elements is not present it will give an error
# so we use discard
my_set.discard(11) 
print(my_set)
```

```python
##pop method
remove=my_set.pop()
print(remove)
print(my_set)
```

```python
##clear all elements
my_set.clear()
print(my_set)
```

```python
##Set Membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
##Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

#Union
union=set1.union(set2)
print(union)

##Intersection
intersection=set1.intersection(set2)
print(intersection)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

##Difference
print(set1.difference(set2))

set1.difference_update(set2)
print(set1)
```

```python
##Symmetric difference
set1.symmetric_difference(set2)
#common elements will be removed and all the elements will be combined
```

```python
##Set Methods
set1={3}
set2={3,4,5} 

## is subset
print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
##to remove the duplicate from sets
lst=[1,2,3,3,4,4,5]
set(lst)
```

```python
##Counting unique words in text

text="In this tutorial we are discussing about sets"
words=text.split()
#conver list of words to set to unique words
print(words)
unique_words=set(words)
print(unique_words)
print(len(unique_words))##length
```
