Tuples are ordered collections of items that areimmutable. They are similar to list,but their immutability makes them them different

```python
##creating
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
##Tuple Operation
numbers=(1,2,3,4)
mixed_tuple=("hi","bye","Swaroop")
concatenation_tuple=numbers+mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple*3
```

```python
numbers*3
```

####Immutable Nature of elements
Tuples are immutable, meaning their elements cannot be changed once assigned 

```python
##Tuple Methods
print(numbers.count(1))
print(numbers.index(3))##returns the index
```

```python
##Packing and Unpacking Tuple
##Packing
packed_tuple=1,"Hello",'c'
print(packed_tuple)
##Unpacking
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
##Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
##nested Tuple

lst=[[1,2,3,4],[6,7,8,9],[1,"heelo",4,'c']]##Nested list
print(lst[0])
lst[0][2]
 
```

```python
nested_tuple=((1,2,3),("a","b","c"),(True,False))
## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
##Iterating over nested tuples
for sub_tuple in nested_tuple:
    for items in sub_tuple:
        print(items,end=" ")
    print()
```
