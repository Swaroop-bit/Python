File Operation

Python providesbuilt-in-function and methods to read from write to files,both text and binary

```python
####Read a whole file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        ##print(line)
        print(line.strip())##removes the newline character
```

```python
## Writing a file(Overwriting)
with open('example.txt','w') as file:
    file.write('Hello World!\n')
```

```python
## Writing a file(without overwriting)
with open('example.txt','a') as file:
    file.write("Append operation taking place!\n")
```

```python
## Writing a list of lines to a file
lines=['First line \n','Second line \n','Third Line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Writing a Binary file

data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## Reading a binary file
with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
##Read the content from a source text file and write to a destination text file
##Copying a text file
with open('example.txt','r') as source_file:
    content=source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)    
```

the w+ mode in python is used to open a file for both reading and writing.If this file doesn't exist,it will be created.if the file escits,its content is truncated(file is overwritten)

```python
## Writing and then reading a file
with open('example.txt','w+') as file:
   file.write("Hello \n")
   file.write("V")
   
   ## Move the file cursor
   file.seek(0)

   content=file.read()
   print(content)
```
