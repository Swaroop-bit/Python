```python
##type conversion
age=25
print(type(age))
 
age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
##Dynamic typing
##python allows the type of a variable change as the program exectues
var=10
print(var,type(var))

var="Hello"
print(var,type(var))

var=3.14
print(var,type(var))
```

```python
##input
age=input("what is the age")
print(age,type(age))
```
