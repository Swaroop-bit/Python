#### Iterator

iterators are advanced Python concepts that allow for eddicient looping and memory management.iterators provide a way to access elements of a collection sequentially without exposing the underlying structure

```python
##Number iterator
my_list=[1,2,3,4,5]
iterator=iter(my_list)

```

```python
try:
    print(next(iterator))
except StopIteration:
    print("Nothing else to iterate")
```

```python
##String Iterator
my_string="Hello"
string_iterator=iter(my_string)

print(next(string_iterator))
print(next(string_iterator))
```
