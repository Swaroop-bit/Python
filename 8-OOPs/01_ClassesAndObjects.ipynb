```python
class Dog:
    ##Constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

##Create Objects
dog1=Dog("Fad",12)
print(dog1)  
print(dog1.name)
print(dog1.age)      
```

```python
##Define a class with instance methods
class Dog:
    def __init__(self,name,dog):
        self.name=name
        self.dog=dog

    def bark(self):
        print(f"{self.name} says woof")

dog1=Dog("Buddy",3)
dog1.bark()
```
