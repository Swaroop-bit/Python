```python
## Method Overiding
class Animal:
    def speak(self):
        return "Sound of the animal"
    
## Derived Class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"    
    
## Derived Class
class Cat(Animal):
    def speak(self):
        return "Meow!"    
##Function that demonstrates polymorphism
        
```

```python
##Polymorphism with Functions and methods

class Shape:
    def area(self):
        return "The area of the figure is not defined"
    
class Reactangle(Shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth   

    def area(self):
        return self.length*self.breadth  

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius   

    def area(self):
        return self.radius*self.radius*3.14

def print_area(shape):
    print(f"The area is {shape.area()}")    

c1=Circle(10)
print_area(c1)
```

####Polymorphism with Abstract bas class

Abstract base class(ABCs) are used to define common methods for groups of a related object.They can enforce that derived class implement particular methods,promoting consistency across different implentation.

```python
from abc import ABC,abstractmethod

##Defineing Abstract Class
class Vehcile(ABC):
    @abstractmethod
    def start_engine(self):
        pass##Function excits but does nothing

#Derived class1
class Car(Vehcile):
    def start_engine(self):
        return "Car engine started"    
```
