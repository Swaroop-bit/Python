####Magic Methods

Magic methods in python, also known as dunder methods(double underscore methods), are special methods that start and end with double underscores.These methods enable you to define the behaviour of objects for built-in operation,such as airthmetic operation,copmarison and more

Already defined in Python adn can be override to change the behavior of the objects

```python
'''
__init__: Initializes a new instance of a class
__str__: Rerurns a string representation of a object
__repr__: Returns a official string representation of a object
__len__: Returns a length of the object
__getitem__: Gets an item from a container
__setitem__: Sets an item in a container
'''
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name}dfs{self.age}adsf"

    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
    

                 
```

```python
person=Person("Krish",32)
print(person)
print(repr(person))
```
