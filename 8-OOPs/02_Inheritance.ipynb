```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype}")    
    
```

```python
car1=Car(4,5,"petrol")
car1.drive()
```

```python
##Single Inheritance
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,selfdriving):
        super().__init__(windows,doors,enginetype)##inheritinf from parent
        self.selfdriving=selfdriving

    def selfDriving(self):
     print(f"Tesla supports sled driving :{self.selfdriving}")    
```

```python
tesla=Tesla(4,5,"electric",True)
tesla.drive()
tesla.selfDriving
```

```python
##Multiple Inheritance

# Base class 1
class Animal:
    def __init__(self, species):
        self.species = species
        print(f"Animal constructor called: species = {self.species}")

    def eat(self):
        print(f"{self.species} is eating")

# Base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner
        print(f"Pet constructor called: owner = {self.owner}")

    def play(self):
        print(f"Pet owned by {self.owner} is playing")

# Derived class inherits from both Animal and Pet
class Dog(Animal, Pet):
    def __init__(self, name, species, owner):
        # Call both parent constructors explicitly
        Animal.__init__(self, species)
        Pet.__init__(self, owner)
        self.name = name
        print(f"Dog constructor called: name = {self.name}")

    def bark(self):
        print(f"{self.name} says woof!")

# Create object of Dog
dog1 = Dog("Buddy", "Dog", "Alice")

# Access methods from both parent classes + its own
dog1.eat()    # from Animal
dog1.play()   # from Pet
dog1.bark()   # from Dog
```
